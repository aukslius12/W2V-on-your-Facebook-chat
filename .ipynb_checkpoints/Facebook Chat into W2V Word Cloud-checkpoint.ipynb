```python
# Libraries.
import pandas as pd
import json
import numpy as np
import gensim
from sklearn.manifold import TSNE
import plotly.graph_objs as go
# Not sure if I need all of this. But then again, I don't try to optimize yet. :)
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 

# Loading from json.
# Path to your facebook messages downloads.
# Path looks like ''...\messages\NameLastname_randomnumbers123\message.json'
path = 'D:\Downloads\messages\PijusJauniskis_666104d042\message.json' # Pijus Jauniskis is my best friend, chat couldn't get more authentic than this.
with open(path) as json_data:
    data = json.load(json_data)
messages = data['messages'] 

# Filtering unnecessary messages.
messages_generic = []
for message in messages:
    if message['type'] == 'Generic':
        messages_generic.append(message)

# Keeping only relevant data, storing into a pandas df.
messages = pd.DataFrame.from_dict(messages_generic)
messages = messages[['content', 'sender_name']]
messages = messages.replace(np.nan, '', regex=True)
```

```python
# Initialising a word dictionary.
sender_name = messages['sender_name'][0]
messages_continuous = {
    'content': [messages['content'][0]],
    'sender_name': [messages['sender_name'][0]]
}
```

```python
# Stores continuous messages that have been typed in different messages (but made by the same person) 
# in a single line.
k = 0
for i in range(1, messages.shape[0]):
    if messages['sender_name'][i] == sender_name:
        messages_continuous['content'][k] = '\n'.join([messages_continuous['content'][k],
             str(messages['content'][i])])
        messages_continuous['sender_name'][k] = messages['sender_name'][i]
    else:
        messages_continuous['content'].append(messages['content'][i])
        messages_continuous['sender_name'].append(messages['sender_name'][i])
        sender_name = messages['sender_name'][i]
        k = k + 1
```

```python
pd.unique(messages_continuous['sender_name']) # Just checking.
```

```python
# Filtering and preprocessing data.
# YOUR NAME BELOW!
name = 'Jurgis Samaitis'

messages_continuous = pd.DataFrame(messages_continuous)
messages_continuous[messages_continuous['sender_name'] == name]
messages_train = messages_continuous['content']

messages_train = [gensim.utils.simple_preprocess(text) for text in messages_train]
```

```python
# Importing the stopwords (ENGLISH).
#from nltk.corpus import stopwords
#stop_words = stopwords.words('english')

# Importing the stopwords (LITHUANIAN).
stop_words = pd.read_json('stopwords-lt.json', encoding='ISO-8859-1')[0].tolist()
```

```python
# Removing the stopwords.
messages_final = []
for sent in messages_train:
    sent_filtered = [w for w in sent if not w in stop_words]
    messages_final.append(sent_filtered)
```

```python
# Training Word2Vec model.
w2v = gensim.models.Word2Vec(messages_final,
                             size=150,
                             window=10,
                             min_count=10,
                             workers=10)            

w2v.train(messages_final, total_examples=len(messages_final), epochs=10)
```

```python
# Reducing dimensionality with t-SNE.
# Creating a vocabulary.
labels = []
tokens = []
for word in w2v.wv.vocab:
    tokens.append(w2v[word])
    labels.append(word)
```

```python
# Training t-SNE.
tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500)
new_values = tsne_model.fit_transform(tokens)
x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])
```

```python
# Visualizing.
init_notebook_mode(connected=True)
plot_data = go.Scatter(x=x,
                       y=y,
                       mode='markers+text',
                       marker = dict(
                           size = 4,
                           color = '#2abdbd'),
                       text=labels,
                       textposition='top',
                       textfont=dict(
                           family='calibri',
                           size=12,
                           color='#4f6978')
                      )

layout = go.Layout(
    xaxis=dict(
        autorange=True,
        showgrid=False,
        zeroline=False,
        showline=False,
        ticks='',
        showticklabels=False
    ),
    yaxis=dict(
        autorange=True,
        showgrid=False,
        zeroline=False,
        showline=False,
        ticks='',
        showticklabels=False
    )
)
fig = go.Figure(data=[plot_data], layout=layout)
iplot(fig)
```

```python
# Plotting the values. INFERIOR. SUPERIOR ABOVE. (This one is faster tho.)
plt.figure(figsize=(16, 16)) 
for i in range(len(x)):
    plt.scatter(x[i], y[i], c='#74dbef')
    plt.annotate(labels[i],
                 xy=(x[i], y[i]),
                 xytext=(5, 2),
                 textcoords='offset points',
                 ha='right',
                 va='bottom')
```
